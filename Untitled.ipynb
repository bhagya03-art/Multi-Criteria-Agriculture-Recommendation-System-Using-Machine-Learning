```python
import pandas as pd

# Load the dataset
df = pd.read_csv("new.csv")

# Merge the 'Crop' and 'Fertilizer' columns into a new column 'Crop_Fertilizer'
df['Crop_Fertilizer'] = df['Crop'] + ', ' + df['Fertilizer']

# Display the first few rows of the updated dataframe
df.head()
```

```python
columns_to_drop = ['Crop', 'Fertilizer']  # replace with the actual column names you want to drop

# Drop the columns
df = df.drop(columns=columns_to_drop)
```
